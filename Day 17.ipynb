## Part 1

```python
import numpy as np
```

```python
input_slices = []

with open('input.txt') as lines:
    wafer = []
    for line in lines:
        wafer.append(line.strip())
    input_slices.append(wafer)
    
def display_slices(slices):
    total = 0
    for x in range(len(slices[0])):
        sub_total = 0
        for i in slices:
            sub_total += i[x].count('#')
            
        total += sub_total
        
        if sub_total > 0:
            for i in slices:
                print(i[x], end=' ')
            print()
    print(total)
    return total

display_slices(input_slices)
slices3d = input_slices[:]
slices4d = [input_slices[:]]
```

```python
neighbours = []
for z in range(3):
    for y in range(3):
        for x in range(3):
            if not ( z == 1 and y == 1 and x == 1 ):
                neighbours.append(np.array([x,y,z]))
```

```python
def convolve(slices):
    output=np.zeros((len(slices)+2, len(slices[0])+2, len(slices[0])+2), dtype=int)
    for z, wafer in enumerate(slices):
        for y, line in enumerate(wafer):
            for x, alive in enumerate(line):
                if alive == '#':
                    for n in neighbours: 
                        output[n[0] + z][n[1] + y][n[2] + x] += 1
    return output

def build_next_state(slices, density_map):
    next_state = []
    for z, wafer in enumerate(density_map):
        next_wafer = []
        for y, line in enumerate(wafer):
            next_line = ''
            for x, alive in enumerate(line):
                if alive < 2 or alive > 3:
                    next_line += '.'
                elif alive == 3:
                    next_line += '#'
                elif alive == 2:
                    if 0 < z < len(density_map)-1 and 0 < y < len(wafer)-1 and 0 < x < len(line)-1:
                        next_line += slices[z-1][y-1][x-1]
                    else:
                        next_line += '.'
            next_wafer.append(next_line)
        next_state.append(next_wafer)
            
    return next_state
            
display_slices(slices3d)
for i in range(6):
    density_map = convolve(slices3d)
    slices3d = build_next_state(slices3d, density_map)

display_slices(slices3d)
```

## Part 2

```python
neighbours_4d = []
for w in range(3):
    for z in range(3):
        for y in range(3):
            for x in range(3):
                if not (w == 1 and z == 1 and y == 1 and x == 1):
                    neighbours_4d.append(np.array([w,x,y,z]))
```

```python
def convolve_4d(cake):
    output=np.zeros((len(cake)+2, len(cake[0])+2, len(cake[0][0])+2, len(cake[0][0])+2), dtype=int)
    for w, slices in enumerate(cake):
        for z, wafer in enumerate(slices):
            for y, line in enumerate(wafer):
                for x, alive in enumerate(line):
                    if alive == '#':
                        for n in neighbours_4d: 
                            output[n[0] + w][n[1] + z][n[2] + y][n[3] + x] += 1
    return output

def build_next_state_4d(cake, density_map):
    next_state = []
    for w, slices in enumerate(density_map):
        next_slices = []
        for z, wafer in enumerate(slices):
            next_wafer = []
            for y, line in enumerate(wafer):
                next_line = ''
                for x, alive in enumerate(line):
                    if alive < 2 or alive > 3:
                        next_line += '.'
                    elif alive == 3:
                        next_line += '#'
                    elif alive == 2:
                        if 0 < w < len(density_map)-1 and 0 < z < len(slices)-1 and 0 < y < len(wafer)-1 and 0 < x < len(line)-1:
                            next_line += cake[w-1][z-1][y-1][x-1]
                        else:
                            next_line += '.'
                next_wafer.append(next_line)
            next_slices.append(next_wafer)
        next_state.append(next_slices)
            
    return next_state
```

```python
slices4d = [input_slices[:]]
for slices in slices4d: display_slices(slices)

for i in range(6):
    density_map = convolve_4d(slices4d)
    slices4d = build_next_state_4d(slices4d, density_map)

total = 0
for slices in slices4d: 
    total += display_slices(slices)
    
print(total)
```
