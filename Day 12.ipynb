## Part 1

```python
import numpy as np
```

```python
position = np.array([0, 0])
cardinals = ['N', 'E', 'S', 'W']
facing = 1 
coordinates = np.array([[0, 1], [1, 0], [0, -1], [-1, 0]])

def change_bearing(facing, letter, amount):
    value = amount // 90
    direction = -1 if letter == 'L' else 1
    facing += value * direction
    if facing < 0:
        return facing + 4
    elif facing >= 4:
        return facing - 4
    return facing

print(position)

with open('input.txt') as lines:
    for instruction in lines:
        instruction = instruction.strip()
        letter = instruction[0]
        amount = int(instruction[1:])
        
        if letter == 'F':
            position += amount * coordinates[facing]
            print("{} * {}: {}".format(instruction, cardinals[facing], position))
        elif letter == 'R' or letter == 'L':
            facing = change_bearing(facing, letter, amount)
            print("{} → {}".format(instruction, cardinals[facing]))
        else:
            position += amount * coordinates[cardinals.index(letter)]
            print("{} * {}: {}".format(instruction, letter, position))
            
print(position)
print(abs(position[0]) + abs(position[1]))
```

## Part 2

```python
sh_position = np.array([0,  0])
wp_position = np.array([10, 1]) # waypoint starts 1 north 10 east

def rotate_waypoint(wp_position, letter, amount):
    value = amount // 90
    turns = value if letter == 'R' else 4 - value
    for i in range(turns):
        new_wp_position = np.zeros(wp_position.shape, dtype=int)
        new_wp_position[0] = wp_position[1]
        new_wp_position[1] = -wp_position[0]
        wp_position = new_wp_position
    
    return wp_position

with open('input.txt') as lines:
    for instruction in lines:
        pass
        instruction = instruction.strip()
        letter = instruction[0]
        amount = int(instruction[1:])
        
        if letter == 'F':
            sh_position += amount * wp_position
            print("\n{} * W{}: {}".format(instruction, wp_position, sh_position))
        elif letter == 'R' or letter == 'L':
            new_wp_position = rotate_waypoint(wp_position, letter, amount)
            wp_position = new_wp_position
        else:
            wp_position += amount * coordinates[cardinals.index(letter)]
            
print(abs(sh_position[0]) + abs(sh_position[1]))
```
