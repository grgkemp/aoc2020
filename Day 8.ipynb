## Part 1

```python
program = []

def load_program(name):
    program = []
    with open(name) as code:
        for line in code:
            instruction = {}
            tokens = line.split(' ')
            instruction['op'] = tokens[0]
            instruction['arg'] = int(tokens[1])
            instruction['exc'] = 0
            program.append(instruction)
    return program
        
print(load_program('input.txt')[:5])
```

```python
accumulator = 0
executing = True
pc = 0
program = load_program('input.txt')

while executing:
    instruction = program[pc]
    
    # check if instruction has been executed before
    if instruction['exc'] > 0:
        executing = False
        print('exit on [{}: {} {}] \naccumulator @ {}'.format(pc, instruction['op'], 
                                                              instruction['arg'], 
                                                              accumulator))
        continue
        
    print('{}: {} {}'.format(pc, instruction['op'], instruction['arg']))
    
    if instruction['op'] == 'acc':
        accumulator += instruction['arg']
        pc += 1
    elif instruction['op'] == 'nop':
        pc += 1
    elif instruction['op'] == 'jmp':
        pc += instruction['arg']
    
    instruction['exc'] += 1
```

## Part 2

```python
for index in range(len(program)):
    program = load_program('input.txt')
        
    if program[index]['op'] == 'acc':
        continue
    else:
        from_i = '[{}: {} {}]'.format(index, program[index]['op'], program[index]['arg'])
        program[index]['op'] = 'nop' if program[index]['op'] == 'jmp' else 'jmp'
        to_i = '[{}: {} {}]'.format(index, program[index]['op'], program[index]['arg'])
        
    accumulator = 0
    pc = 0
    executing = True
    
    while executing:
        # check if EOF reached
        if pc >= len(program):
            executing = False
            print('{} ⟼ {}: finish on [{}: EOF] accumulator @ {}'.format(from_i, to_i, pc, accumulator))
            break

        instruction = program[pc]
        
        # check if instruction has been executed before
        if instruction['exc'] > 0:
            print('{} ⟼ {}: err on [{}: {} {}] accumulator @ {}'.format(from_i, to_i, pc, instruction['op'], 
                                                                  instruction['arg'], accumulator))
            break

        if instruction['op'] == 'acc':
            accumulator += instruction['arg']
            pc += 1
        elif instruction['op'] == 'nop':
            pc += 1
        elif instruction['op'] == 'jmp':
            pc += instruction['arg']

        instruction['exc'] += 1
    
    if not executing: break
```
