## Part 1

```python
timestamp = 0
bus_times = []

with open('input.txt') as lines:
    full_input = lines.readlines()
    timestamp = int(full_input[0])
    bus_ids = full_input[1].split(',')
    for bus in bus_ids:
        if bus.isdigit(): bus_times.append(int(bus))
            
print(timestamp)
print(bus_times)
```

```python
def time_to_wait(timestamp, bus_id):
    return (bus_id - timestamp) % bus_id

min_wait = 1000
min_id = None
for bus_id in bus_times:
    wait = time_to_wait(timestamp, bus_id)
    print('id: {}, {}'.format(bus_id, wait))
    if wait < min_wait:
        min_wait = wait
        min_id = bus_id
        
print('{} * {} = {}'.format(min_id, min_wait, min_id * min_wait))
```

```python
timestamp = 0
bus_ids = None

with open('input.txt') as lines:
    full_input = lines.readlines()
    timestamp = int(full_input[0])
    bus_ids = full_input[1].split(',')
```

```python
def display_timetable(timestamp, bus_ids):
    print(" "*(len(str(timestamp))-2), end='')
    print("ts|", end='')
    line = 1
    for b_id in bus_ids: 
        if b_id != 'x': 
            print("{:6d}|".format(int(b_id)), end='')
            line += 1
    print()
    print('=' * ((7 * line) + len(str(timestamp)))) 
    for ts in range(timestamp, timestamp + len(bus_ids)):
        print("{:6d}|".format(ts), end='')
        for b_id in bus_ids:
            if b_id == 'x':
                pass
            else:
                print("{:6d}|".format(ts % int(b_id)), end = '')
        print()
```

```python
import math
# thanks wikipedia

def find_bezout_i(a, b):
    old_r, r = (a, b)
    old_s, s = (1, 0)
    old_t, t = (0, 1)
    
    while not r == 0:
        quotient = old_r // r
        old_r, r = (r, old_r - quotient * r)
        old_s, s = (s, old_s - quotient * s)
        old_t, t = (t, old_t - quotient * t)
    
    return (old_s, old_t)

def calculate(buses):
    n1 = int(buses[0])
    a1 = 0
    n2, a2, m1, m2 = None, None, None, None
    for i, bus in enumerate(buses[1:]):
        if bus == 'x':
            continue
        n2 = int(bus)
        a2 = -(i + 1)
        m1, m2 = find_bezout_i(n1, n2)
        x1 = a1*m2*n2 + a2*m1*n1
        x1 = x1 % (n1*n2)
        a1 = x1
        n1 = n1*n2
        
    return a1

timestamp = calculate(bus_ids)
display_timetable(timestamp, bus_ids)
```
